```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
from PIL import Image
```

```python
img = Image.open('./testNoise.jpg').convert('L')
```

```python
np_img  = np.asarray(img)
```

```python
print(np_img.shape)
print(np.amax(np_img))
print(np.amin(np_img))
```

```python
plt.imshow(np_img)
```

```python
hist_arr = np.zeros((np.amax(np_img)+1, 1))
print(hist_arr.shape)
```

```python
np_img[111][111]
```

```python
for r in range(0, np_img.shape[0]):
    for c in range(0, np_img.shape[1]):
        #print(r, c, np_img[r][c])
        hist_arr[np_img[r][c]][0] += 1
```

```python
plt.figure(figsize=(20,10))
plt.bar(range(hist_arr[:,0].shape[0]), list(hist_arr[:,0].astype('int64')))
plt.show()
```

```python
import seaborn as sns
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(100, 40)
sns.barplot(list(range(hist_arr[:,0].shape[0])), list(hist_arr[:,0].astype('int64')))
fig.savefig('example.png')
```
