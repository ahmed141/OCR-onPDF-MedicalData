```python
start = time.time()

with PyTessBaseAPI(path = "C:\\Program Files\\Tesseract-OCR\\tessdata") as api:
    api.SetImageFile('../datafiles/Page_2.jpg')
    api.SetVariable("save_blob_choices", "T")
    api.SetVariable('hocr_font_info', '1')
    api.SetVariable('tessedit_debug_fonts', '1')
    api.Recognize()
    pi = api.AnalyseLayout()
    ri = api.GetIterator()
    
    print(pi)
    print(ri)
    
    print("FONT:\t", ri.WordFontAttributes())
    
    level = RIL.BLOCK
    for r in iterate_level(ri, level):
        print(r)
        print("font:\t", r.WordFontAttributes(), r.GetUTF8Text(level))
        # print(ri.)
    print('--------------------------------------------- BLOCK start')
    ri = api.GetIterator()
    level = RIL.PARA
    for r in iterate_level(ri, level):
        print(r)
        print("font:\t", r.WordFontAttributes())
        # print(ri.)
    print('--------------------------------------------- PARA start')
    ri = api.GetIterator()
    level = RIL.TEXTLINE
    for r in iterate_level(ri, level):
        print(r)
        print("font:\t", r.WordFontAttributes())
        # print(ri.)
    print('--------------------------------------------- TEXTLINE start')
    ri = api.GetIterator()
    level = RIL.WORD
    for r in iterate_level(ri, level):
        print(r)
        print("font:\t", r.WordFontAttributes())
        # print(ri.)
    print('--------------------------------------------- WORD start')
    ri = api.GetIterator()
    level = RIL.SYMBOL
    for r in iterate_level(ri, level):
        print(r)
        print("font:\t", r.WordFontAttributes())
        # print(ri.)
    print('--------------------------------------------- SYMBOL start')

print("TIME:\t", time.time() - start)
```

```python
with PyTessBaseAPI() as api:
    api.SetImage(imageMAT)
    boxes = api.GetComponentImages(RIL.TEXTLINE, True)
    print('Found {} textline image components.'.format(len(boxes)))
    for i, (im, box, _, _) in enumerate(boxes):
        # im is a PIL image object
        # box is a dict with x, y, w and h keys
        api.SetRectangle(box['x'], box['y'], box['w'], box['h'])
        ocrResult = api.GetUTF8Text()
        conf = api.MeanTextConf()
        print(u"Box[{0}]: x={x}, y={y}, w={w}, h={h}, "
              "confidence: {1}, text: {2}".format(i, conf, ocrResult, **box))
```
